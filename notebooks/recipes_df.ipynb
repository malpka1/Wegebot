```python
# Usunięcie powtarzających się słów w każdej komórce kolumny
recipes_df["ingredients_lemmatized"] = recipes_df["ingredients_lemmatized"].apply(lambda x: " ".join(sorted(set(x.split()), key=x.split().index)))

recipes_df.head(20)
```
